```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use('fivethirtyeight')

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
#Read in both player totals and allshots
player_totals = pd.read_csv('/Users/ericnesi/Desktop/capstone_eric/datasets/NBL_Player_Data.csv')
allshots = pd.read_csv('/Users/ericnesi/Desktop/capstone_eric/datasets/All_Shots.csv', index_col=[0])
```

```python
print(player_totals.head())
print(allshots.head())
```

```python
#clean name up
player_totals['First_Init'] = player_totals['Player'].map(lambda x: str(x)[:1])
player_totals[['First_Name','Second_Name', 'Third_Name']]= pd.DataFrame([ x.split() for x in player_totals['Player'].tolist()])
```

```python
player_totals.head()
```

```python
#continue to clean name
player_totals['Third_Name'] = player_totals['Third_Name'].astype(str)
player_totals.loc[player_totals['Third_Name'] == 'None', 'Third_Name'] = ''
```

```python
player_totals.head(4)
```

```python
#help me to concatenate by replacing none with empty space
player_totals.loc[player_totals['Third_Name'] == None, 'Third_Name'] = ''
```

```python
player_totals.head()
```

```python
#concatenate all names
player_totals['player'] = player_totals['First_Init'] +'.'+ ' ' + player_totals['Second_Name'] + player_totals['Third_Name'] 
```

```python
#drop repeat columns
player_totals = player_totals.drop(['First_Name', 'Player', 'Second_Name', 'Third_Name'], axis=1)
```

```python
player_totals.head()
```

```python
#make sure name is a str
allshots['player'] = allshots['player'].astype(str)
player_totals['player'] = player_totals['player'].astype(str)
```

```python
#separate df to merge
player_names = player_totals[['player', 'Team']].copy()
```

```python
#commence merge
allshots = allshots.merge(player_names, on='player', how='left')
```

```python
allshots.head()
```

```python
#lots of nulls
allshots.isnull().sum().sort_values(ascending=False).head(5)
```

```python
#rename for column names to match my Kings column names
allshots = allshots.rename(columns={'Team':'Team', 'player':'Player', 'shirtNumber':'PlayerNo', 'x': 'X', 'y':'Y', 'actionType': 'ShotType'})
```

```python
allshots.head()
```

```python
#rename continued, reorder
allshots = allshots.rename(columns={'r':'Result', 'subType': 'SubType', 'per': 'Quarter'})
allshots = allshots[['Team', 'Player', 'PlayerNo', 'X', 'Y', 'ShotType', 'SubType', 'Result', 'Quarter']]
```
